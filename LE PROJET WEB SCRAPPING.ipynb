# PROJET WEB SCRAPING

## Définition du Web Scraping
Le web scraping est une technique utilisée pour extraire automatiquement des informations à partir de sites web. Cela se fait généralement en téléchargeant la page web et en analysant son contenu pour en extraire les données souhaitées. Cette technique utilise des scripts ou des logiciels qui envoient des requêtes HTTP aux sites web et traitent ensuite le HTML reçu pour obtenir les informations pertinentes.


## Présentation du Site Expat-Dakar
Expat-Dakar est une plateforme en ligne populaire au Sénégal, dédiée à la mise en relation des expatriés et des locaux avec divers services, notamment la location de voitures. Le site propose une large gamme de véhicules à louer, allant des voitures économiques aux véhicules de luxe, en passant par les utilitaires. Les utilisateurs peuvent rechercher des voitures en fonction de leurs besoins spécifiques, comparer les prix, et contacter directement les propriétaires ou les agences de location.


## Problématique
Lorsqu'on envisage de réaliser du web scraping sur un site comme Expat-Dakar, plusieurs problématiques peuvent survenir :

Accès et Permissions :

Les conditions d'utilisation du site peuvent interdire le scraping. Il est crucial de vérifier les termes de service pour éviter des problèmes légaux.
Complexité des Données :

Les données sur les sites web peuvent être imbriquées dans des structures HTML complexes, rendant leur extraction difficile.
Changements de Structure :

Les sites web évoluent régulièrement. Les changements dans la structure HTML peuvent casser les scripts de scraping existants.
Gestion des Requêtes :

Envoyer trop de requêtes en peu de temps peut être perçu comme une attaque par déni de service, entraînant un blocage par le site cible.

### Définition d'une fonction pour automatiser l'extraction ou l'importation de la page du site web que nous allons attaquer 

```python
def extraire_doc(notre_url):
    #topic_url = 'https://www.expat-dakar.com/voitures/nissan'
    response = requests.get(notre_url)
    type(response)
    page_contents = response.text
    taille_de_la_page=len(page_contents)
    with open('scraper.html', 'w', encoding="utf-8") as file:
        file.write(page_contents)
        #scraper=file.read()
    doc = BeautifulSoup(page_contents, 'html.parser')
    return (doc)    
```

# Colonne numéro de téléphone

```python
# Trouver toutes les balises <a> avec la classe spécifique
def extraire_telephone(doc):
    
    phone_tags = doc.find_all('a', class_='listing-card__contact__number-inner')

    # Liste pour stocker les numéros de téléphone
    phone_numbers = []

    # Extraire les numéros de téléphone de chaque balise <a>
    for tag in phone_tags:
        phone_number = tag.text.strip()
        phone_numbers.append(phone_number)

    return (phone_numbers)
```

### Ainsi comme tout bon progammeur et en tant que professionnels de la data nous allons automatisé toutes ces taches  pour faciliter l'extraction d'informations sur d'autre pages ou d'autres types d'articles dans notre site 

```python
def extraire_la_base_de_donnee(soup):
    # Trouver toutes les balises <a> avec la classe spécifique
    phone_tags = soup.find_all('a', class_='listing-card__contact__number-inner')

    # Liste pour stocker les informations
    listings = []

    # Extraire les informations de chaque balise <a>
    for tag in phone_tags:
        phone_number = tag.text.strip()
        car_title = tag.get('data-t-listing_title', 'N/A')
        car_price = tag.get('data-t-listing_price', 'N/A')
        car_location = tag.get('data-t-listing_location_title', 'N/A')
        listings.append({
            "phone_number": phone_number,
            "car_title": car_title,
            "car_price": car_price,
            "car_location": car_location
        })

    # Afficher les informations extraites
    for listing in listings:
        print(f"Numéro de téléphone: {listing['phone_number']}")
        print(f"Marque de voiture: {listing['car_title']}")
        print(f"Prix: {listing['car_price']}")
        print(f"Emplacement: {listing['car_location']}")
        print("-" * 40)
        
```

```python

import csv

def extraire_la_base_de_donnee(doc):
    # Trouver toutes les balises <a> avec la classe spécifique pour les numéros de téléphone
    phone_tags = doc.find_all('a', class_='listing-card__contact__number-inner')

    # Liste pour stocker les informations
    listings = []

    # Extraire les informations de chaque balise <a>
    for tag in phone_tags:
        phone_number = tag.text.strip()
        car_title = tag.get('data-t-listing_title', 'N/A')
        car_price = tag.get('data-t-listing_price', 'N/A')
        car_location = tag.get('data-t-listing_location_title', 'N/A')
        car_category = tag.get('data-t-listing_category_title', 'N/A')

        listings.append({
            "phone_number": phone_number,
            "car_title": car_title,
            "car_price": car_price,
            "car_location": car_location,
            "car_category": car_category
        })

    return listings

```

```python
def save_to_csv(data, filename):
    # Définir les en-têtes du CSV
    headers = ["phone_number", "car_title", "car_price", "car_location", "car_category"]

    # Écrire les données dans un fichier CSV
    with open(filename, 'w', newline='', encoding='utf-8') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=headers)
        writer.writeheader()
        writer.writerows(data)
```
